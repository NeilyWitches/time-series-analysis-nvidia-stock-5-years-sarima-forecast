```r
nvidia <- read.csv('C:/Users/drumm/Documents/practical time series analysis/nvidiastockprice5yrs.csv')
prices1 <- nvidia$Close.Last
prices2 <- rev(prices1)
prices3 <- as.character(prices2)
prices4 <- gsub(' \\$', '', prices3)
prices5 <- as.numeric(prices4)
prices6 <- ts(prices5)
```

```r
jpeg('pricestransformed.jpg')
par(mfrow=c(3,1))

plot(prices6, main='Daily sales for NVIDIA stock at market close', ylab='', col='blue', lwd=3)
plot(log(prices6), main='Log-transform of stock prices', ylab='', col='red', lwd=3)
plot(diff(log(prices6)), main='Differenced Log-transform of stock prices', ylab='', col='green', lwd=3)
dev.off()
```

```r
data<-diff(log(prices6))
jpeg('acfandpacftransformed.jpg')
acf2(data,50)
dev.off()
```

```r
d=1
for(p in 1:2){
  for(q in 1:2){
        if(p+d+q<=10){
          model<-arima(x=log(prices6), order = c((p-1),d,(q-1)))
          pval<-Box.test(model$residuals, lag=log(length(model$residuals)))
          sse<-sum(model$residuals^2)
          cat(p-1,d,q-1, 'AIC=', model$aic, ' SSE=',sse,' p-VALUE=', pval$p.value,'\n')
        }
      }
    }
```

```r
model<- arima(x=log(prices6), order = c(0,1,0))
jpeg('forecastdays.jpg')
plot(forecast(model))
dev.off()
```
